```python
import pandas as pd
import numpy as np

def readData(filename,columns):
    data = pd.read_csv(filename, sep="\t",header=None)
    data.columns = columns
    return data

expert_data = readData('user_info.txt',["U_id","U_tag","U_Word_seq","U_Char_seq"])
question_data = readData('question_info.txt',["Q_id","Q_tag","Word_seq","Q_Char_seq","Likes","Ans","Top_Ans"])
expert_question_dist = readData('invited_info_train.txt',["Q_id","U_id","output"])
train_data_matrix = pd.DataFrame(0.0,index = expert_data["U_id"],columns = question_data["Q_id"])
for i,row in expert_question_dist.iterrows():
    train_data_matrix[row["Q_id"]][row["U_id"]] = row["output"]
    
```

```python
import scipy.sparse as sp
from scipy.sparse.linalg import svds
#run for different values of K and find which one is better in training data
u, s, vt = svds(train_data_matrix, k = 15)
s_diag_matrix=np.diag(s)
X_pred = np.dot(np.dot(u, s_diag_matrix), vt)
pd.DataFrame(X_pred,index = expert_data["U_id"],columns = question_data["Q_id"]) 
```

```python
X_pred = pd.DataFrame(X_pred,index = expert_data["U_id"],columns = question_data["Q_id"]) 
validation_data = pd.read_csv('validate_nolabel.txt', sep=",")[[0,1]]
validation_data.columns = ["Q_id","U_id"]
validation_data["label"] = ""
for i,row in validation_data.iterrows():
     row["label"]=X_pred[row["Q_id"]][row["U_id"]]
validation_data.to_csv('val_expert.csv',sep=',', index=False, header = False)
```
